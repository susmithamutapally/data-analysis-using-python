#### A. Nested Data Structures and Lists

The students received their grades for a midterm exam. The results came as a dictionary, and the grades are in the form of lists. The list contains the score each student received, for each question in the midterm. Unfortunately, the professor was very absent minded and he forgot to sum up the points to compute the overall grade for each student.

* How many students are in the data set?
* How many questions were in the midterm?
* Write code to compute the overall grade of each student. Add the overall grade in a new dictionary called `midterm_grades`, with the students as keys, and the overall grades as (integer) values.
* Extract the values from the `midterm_grades` and create a list of the overall grades for the midterm.
* Compute the maximum, minimum, average, and median grade among the overall student grades.


```python
grades = {
    "studentA" : [5, 2, 3, 1, 2, 4, 5, 5, 1, 1],
    "studentB" : [1, 2, 5, 5, 2, 0, 1, 5, 5, 2],
    "studentC" : [2, 5, 5, 5, 5, 0, 1, 3, 1, 0],
    "studentD" : [0, 1, 0, 5, 2, 5, 1, 3, 3, 5],
    "studentE" : [5, 4, 5, 5, 1, 5, 1, 4, 4, 5],
    "studentF" : [2, 2, 1, 1, 1, 3, 1, 3, 0, 1],
    "studentG" : [5, 5, 5, 5, 5, 4, 3, 5, 3, 1],
}

# How many students are in the data set?

number_of_students = len(grades)
print("Number of students:", number_of_students)

# How many questions were in the midterm?

number_of_questions = len(grades["studentA"])
print("Number of questions in the midterm:", number_of_questions)

# Compute the overall grade of each student

midterm_grades = {student:sum(scores) for (student,scores) in grades.items()}
print("Overall grades of each student:",midterm_grades)

# Extract the values from the midterm_grades and create a list of the overall grades for the midterm.

grades_list = [scores for scores in midterm_grades.values()]
print("List of overall grades:",grades_list)

max_grade = max(grades_list)
print("Maximum grade:",max_grade)


min_grade = min(grades_list)
print("Minimum grade:",min_grade)

average_grade = sum(grades_list) / len(grades_list)
print("Average grade:",average_grade)

grades_list.sort()
mid = len(grades_list) // 2
res = (grades_list[mid] + grades_list[~mid]) / 2

print("Median of list is : " + str(res))
```

#### C. Function that returns a dictionary 

Write a function that has a parameter that is the height of someone in centimeters. You need to convert it into feet and inches. (Each foot is 30.48 cm. Each inch is 2.54cm.). The function should return the height in feet and inches. The result should be a dictionary with two keys, `feet` and `inches`, and the values for both keys should be integers.

```python
def convert_height(height_cm):
    height_inches = height_cm / 2.54
    total_feet = height_inches // 12
    remaining_inches = height_inches % 12
    
    height_dict = {
        'feet': int(total_feet),
        'inches': int(remaining_inches)
    }
    
    return height_dict

print("The result dictionary is:", convert_height(160))

```
